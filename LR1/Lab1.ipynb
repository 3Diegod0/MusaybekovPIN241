#6. С помощью groupby() и agg() определяем:
•какие команды были самыми результативными (сумма голов)
самыми точными (средний % точности ударов)

•какие команды совершили больше всего фолов, получили больше
всего желтых и красных карточек

•насколько равномерно использовался состав. Посчитайте среднее и
стандартное отклонение по количеству использованных игроков
для всех команд

•ТОП-5 команд по разным показателям: общему количеству ударов
в створ, лучшему соотношению сэйвов к ударам, худшей
дисциплине (сумма желтых и красных карточек)

```python
for col in ['Shooting Accuracy', '% Goals-to-shots', 'Saves-to-shots ratio']:
    if col in df.columns and df[col].dtype == 'object':  # только если это строка
        df[col] = df[col].str.rstrip('%').astype(float)
    elif col in df.columns and df[col].dtype != 'float64':
        # если уже число, но не float
        df[col] = df[col].astype(float)

# 1. Самые результативные (сумма голов) и самые точные (средний % точности)
result_stats = df.groupby('Team').agg({
    'Goals': 'sum',
    'Shooting Accuracy': 'mean'
}).sort_values('Goals', ascending=False)

print("1. САМЫЕ РЕЗУЛЬТАТИВНЫЕ И САМЫЕ ТОЧНЫЕ КОМАНДЫ")
print(result_stats)
print("\n Вывод: Испания — лучшая по голам (12) и точности (55.9%).\n")

# 2. Команды, совершившие больше всего фолов, желтых и красных карточек
discipline_stats = df.groupby('Team').agg({
    'Fouls Conceded': 'sum',
    'Yellow Cards': 'sum',
    'Red Cards': 'sum'
}).sort_values('Yellow Cards', ascending=False)

print("2. ДИСЦИПЛИНА (ФОЛЫ, ЖЁЛТЫЕ И КРАСНЫЕ КАРТОЧКИ)")
print(discipline_stats)
print("\n Вывод: Италия — больше всех жёлтых (16), Португалия — больше фолов (90).\n")

# 3. Равномерность использования состава (среднее и std по Players Used)
usage_mean = df['Players Used'].mean()
usage_std = df['Players Used'].std()

print("3. РАВНОМЕРНОСТЬ ИСПОЛЬЗОВАНИЯ СОСТАВА")
print(f"Среднее количество использованных игроков: {usage_mean:.2f}")
print(f"Стандартное отклонение: {usage_std:.2f}")
print("\nВывод: Состав используется равномерно (разброс всего 1.5 игрока).\n")

# 4. ТОП-5 по разным показателям
# 4.1 Общее количество ударов в створ
top_shots = df.nlargest(5, 'Shots on target')[['Team', 'Shots on target']]

# 4.2 Лучшее соотношение сэйвов к ударам
top_saves = df.nlargest(5, 'Saves-to-shots ratio')[['Team', 'Saves-to-shots ratio']]

# 4.3 Худшая дисциплина (сумма желтых и красных)
df['Total Cards'] = df['Yellow Cards'] + df['Red Cards']
top_cards = df.nlargest(5, 'Total Cards')[['Team', 'Yellow Cards', 'Red Cards', 'Total Cards']]


print("4. ТОП-5 КОМАНД ПО РАЗНЫМ ПОКАЗАТЕЛЯМ")
print("\n ТОП-5 по ударам в створ:")
print(top_shots)
print("\n ТОП-5 по % сэйвов:")
print(top_saves)
print("\n ТОП-5 по худшей дисциплине (жёлтые + красные):")
print(top_cards)

# Удалим временный столбец, чтобы не мешал
df.drop('Total Cards', axis=1, inplace=True)
```

#7.Разобьём команды на группы по количеству использованных игроков. Для каждой группы посчитаем среднее количество голов.

```python
bins = [11, 15, 20, 23]
labels = ['11-15', '16-20', '21-23']
df['PlayersGroup'] = pd.cut(df['Players Used'], bins=bins, labels=labels, right=False)

# Среднее количество голов по группам
grouped_goals = df.groupby('PlayersGroup')['Goals'].mean()
print(grouped_goals)
```

#8.С помощью pivot_table() построим сводные таблицы, показывающие:
• какие команды были эффективнее в завершении, где строка –
команда, а столбцы – средние значения: goals-to-shots, shooting
accuracy, hit woodwork

•указание на атакующий футбол с длинными передачами,
группировка команд по их стилю: много ли они выигрывают фолов
и при этом часто попадают в офсайд

•эффективность реализации пенальти, где отображаются: команды,
количество назначенных пенальти (penalty goals + penalties not
scored), процент реализации (penalty goals / общее кол-во * 100%).

```python
df['penalties_total'] = df['Penalty goals'] + df['Penalties not scored']
df['penalty_accuracy'] = (df['Penalty goals'] / df['penalties_total'].replace(0, float('nan'))) * 100


print("1. ЭФФЕКТИВНОСТЬ ЗАВЕРШЕНИЯ АТАК")


pivot_efficiency = df.pivot_table(
    index='Team',
    values=['% Goals-to-shots', 'Shooting Accuracy', 'Hit Woodwork'],
    aggfunc='mean'
).sort_values('% Goals-to-shots', ascending=False)

print(pivot_efficiency.round(1).to_string())
print("\nВывод: Дания показала лучшую эффективность (20% голов от ударов), Испания - лучшую точность (55.9%), Португалия чаще других попадала в штангу/перекладину (6 раз).")
print("\n")


print("2. АТАКУЮЩИЙ СТИЛЬ: ВЫИГРАННЫЕ ФОЛЫ И ОФСАЙДЫ")


pivot_style = df.pivot_table(
    index='Team',
    values=['Fouls Won', 'Offsides'],
    aggfunc='mean'
).sort_values('Fouls Won', ascending=False)

print(pivot_style.round(1).to_string())
print("\nВывод: Италия выиграла больше всех фолов (101), Испания чаще всех попадала в офсайд (19 раз) - это указывает на атакующий стиль с высоким прессингом и длинными передачами.")
print("\n")



print("3. ЭФФЕКТИВНОСТЬ РЕАЛИЗАЦИИ ПЕНАЛЬТИ")


pivot_penalties = df.pivot_table(
    index='Team',
    values=['penalties_total', 'penalty_accuracy'],
    aggfunc='mean'
)

# Оставляем только команды, у которых были пенальти
pivot_penalties = pivot_penalties[pivot_penalties['penalties_total'] > 0].sort_values('penalty_accuracy', ascending=False)

print(pivot_penalties.round(1).to_string())
print("\nВывод: Германия - единственная команда, реализовавшая 100% пенальти (1 из 1). Греция и Испания получили по 2 пенальти, но реализовали только 50%.")
print("\n")


df.drop(['penalties_total', 'penalty_accuracy'], axis=1, inplace=True)


```
