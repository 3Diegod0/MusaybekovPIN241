- **Team** – название команды;
- **Goals** – забитые голы;
- **Shots on target** – удары в створ ворот;
- **Shots off target** – удары мимо ворот;
- **Shooting Accuracy** – точность ударов;
- **% Goals-to-shots** – процент голов от ударов;
- **Total shots (inc. Blocked)** – всего ударов (включая заблокированные);
- **Hit Woodwork** – попадания в штангу/перекладину;
- **Penalty goals** – голы с пенальти;
- **Penalties not scored** – нереализованные пенальти;
- **Headed goals** – голы головой;
- **Passes** – пасы;
- **Passes completed** – точные пасы;
- **Passing Accuracy** – точность пасов;
- **Touches** – касания мяча;
- **Crosses** – навесы;
- **Dribbles** – обводки;
- **Corners Taken** – угловые;
- **Tackles** – отборы;
- **Clearances** – выносы мяча;
- **Interceptions** – перехваты;
- **Clearances off line** – выносы с линии ворот;
- **Clean Sheets** – "сухие" матчи;
- **Blocks** – блокировки ударов;
- **Goals conceded** – пропущенные голы;
- **Saves made** – сейвы;
- **Saves-to-shots ratio** – процент сейвов;
- **Fouls Won** – заработанные фолы;
- **Fouls Conceded** – совершенные фолы;
- **Offsides** – офсайды;
- **Yellow Cards** – желтые карточки;
- **Red Cards** – красные карточки;
- **Subs on** – замены на выход;
- **Subs off** – замены на вход;
- **Players Used** – использованные игроки.

Наибольшее количество голов (12) зафиксировано у команды с 9 желтыми карточками — это Испания. В целом, можно заметить, что команды с большим количеством желтых карточек (9-10) показывают более низкие результаты по голам по сравнению с командами, у которых желтых карточек меньше (6-7).

Большинство команд сконцентрировано в диапазоне ударов в створ 5-15 и голов 2-6. Также заметны отдельные точки-выбросы с высокими показателями: Испания (22 удара в створ, 12 голов) и Германия (32 удара в створ, 10 голов).

Заметно, что Испания (22 удара в створ, 12 голов) и Германия (32 удара в створ, 10 голов) — команды без красных карточек, что показывает, что высокая результативность не обязательно связана с грубой игрой.

# 7. Корреляция признаков

В реальных задачах машинного обучения при первичном анализе данных необходимо выявить корреляции признаков обучающей выборки. В пакете Pandas имеется встроенный инструмент для этого – метод corr() класса DataFrame. Ниже показан фрагмент вывода этой функции.

```python
data.corr(numeric_only=True)
```

> Матрица корреляции имеет размер 30×30 (по числу числовых признаков). Видно, что некоторые признаки сильно коррелируют между собой.Высокая корреляция означает, что эти признаки тесно связаны и фактически дублируют друг друга.

Полученная матрица корреляции достаточно большая, и анализировать её вручную трудоёмко. Лучше использовать специальный тип графика — heatmap

```python
sns.heatmap(data.corr(numeric_only=True), cmap=plt.cm.PuBuGn)
```

> На тепловой карте корреляции хорошо видны тёмные области — это пары признаков с сильной корреляцией. Наиболее тёмные ячейки наблюдаются между Passes и Touches (0.99), а также между Goals, Shots on target и Shots off target. Также заметна тёмная область отрицательной корреляции у Goals conceded с другими признаками (светлые ячейки). Признак Red Cards выделяется на карте — его строка и столбец практически однородного светлого цвета, что подтверждает отсутствие сильной связи с другими признаками.

Коррелирующие признаки обычно удаляются и не рассматриваются в процессе обучения. Из карты heatmap видно, что некоторые признаки коррелируют: например, сильная корреляция в парах (Goals, Shots on target), (Goals, Shots off target). Из таких пар можно удалить один признак

```python
data_uncorr = data.drop(feats, axis=1)
data_uncorr.columns
```

>После удаления коррелирующих признаков (Goals, Shots on target, Shots off target, Passes) в датафрейме осталось 28 столбцов. Это позволяет снизить мультиколлинеарность при построении моделей машинного обучения.

Перестраиваем heatmap без коррелирующих признаков

```python
sns.heatmap(data_uncorr.corr(numeric_only=True), cmap=plt.cm.PuBuGn)
```

> На обновлённой тепловой карте после удаления коррелирующих признаков видно, что сильных корреляций между оставшимися признаками стало значительно меньше. Наиболее заметная связь — корреляция между Yellow Cards и Fouls Conceded (около 0.6): чем больше фолов совершает команда, тем больше жёлтых карточек она получает. Также видна положительная корреляция между Goals conceded и Saves made — команды, которые пропускают много голов, делают больше сейвов. Признак Red Cards по-прежнему не имеет сильных корреляций с другими признаками.
